###Demonstration of One2onePlot class for evaluating measured vs. modelled results

```python
import sys
sys.path.append('../pestools')
from res import *

% matplotlib inline

res = Res('../cc/Columbia.res')
```

####plot one group using default settings

```python
res.plot_one2one('WCRs1')
```

####plot multiple groups using default settings; add statistics summary

```python
fig, ax = res.plot_one2one(['WCRs2', 'WCRs1'], print_stats=['Mean', 'AME', 'RMSE'])
```

####plot multiple groups with customized settings via a dictionary of Matplotlib keyword arguments
####reduce the legend fontsize by adjusting the rcParam

```python
groupinfo_heads = {'head_best': {'marker': '^',
                                 's': 120,
                                 'c': 'r',
                                 'label': 'Heads, best',
                                 'zorder': 20},
                   'head_good': {'marker': 's',
                                 's': 60,
                                 'c': 'b',
                                 'label': 'Heads, good',
                                 'zorder': 19},
                   'head_fair': {'marker': 'o',
                                 's': 28,
                                 'c': '0.85',
                                 'label': 'Heads, fair',
                                 'zorder': 8,
                                 'alpha': 0.5},
                   'head_poor': {'marker': '+',
                                 's': 20,
                                 'c': 'k',
                                 'label': 'Heads, zero-weighted',
                                 'zorder': 7},
                   'head_zero': {'marker': '+',
                                 's': 20,
                                 'c': 'k',
                                 'label': 'Heads, zero-weighted',
                                 'zorder': 7},
                   'WCRs1': {'marker': 'o',
                             's': 28,
                             'c': '0.85',
                             'label': 'Heads, fair',
                             'zorder': 8,
                             'alpha': 0.5},
                   'WCRs2': {'marker': '+',
                             's': 20,
                             'c': 'k',
                             'label': 'Heads, zero-weighted',
                             'zorder': 7}
}
plt.rcParams['legend.fontsize'] = 8
fig, ax = res.plot_one2one(groupinfo_heads, print_stats=['Mean', 'MAE', 'RMSE'])
fig.savefig('test_one2one.pdf')
```

####HexbinPlot subclass of One2onePlot class

```python
fig, ax = res.plot_hexbin(['head_fair', 'wcrs1'], title='Fair Heads', print_stats=['Mean', 'MAE', 'RMSE'])
```

####Change the colormap and tweak the line using keyword arguments to pyplot

```python
res.plot_hexbin(['head_fair', 'wcrs1'], line_kwds={'linewidth': 4}, cmap='YlOrRd_r')
```

####Override the log bin count default

```python
res.plot_hexbin(['head_fair', 'wcrs1'], bins=None)
```
